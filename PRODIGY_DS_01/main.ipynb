### BAR CHART

```python
import matplotlib.pyplot as plt

# Sample data
genders = ['Male', 'Female', 'Other']
counts = [40, 30, 30]

# Create a figure and axis
fig, ax = plt.subplots()

# Create a bar chart
ax.bar(genders, counts)

# Set title and labels
ax.set_title('Distribution of Genders')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')

# Show the plot
plt.show()
```

### HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
ages = np.random.normal(30, 10, 100)  # Generate 100 random ages with a mean of 30 and standard deviation of 10

# Create a figure and axis
fig, ax = plt.subplots()

# Create a histogram
ax.hist(ages, bins=10, edgecolor='black')

# Set title and labels
ax.set_title('Distribution of Ages')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()
```

### STACKED BAR CHART

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x1 = np.random.rand(10)
x2 = np.random.rand(10)
y = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Create a figure and axis
fig, ax = plt.subplots()

# Create a stacked bar chart
ax.bar(y, x1, label='Group 1')
ax.bar(y, x2, bottom=x1, label='Group 2')

# Set title and labels
ax.set_title('Stacked Bar Chart Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.show()
```

### GROUPED BAR CHART

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x1 = np.random.rand(10)
x2 = np.random.rand(10)
y = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Create a figure and axis
fig, ax = plt.subplots()

# Create a grouped bar chart
bar_width = 0.35
ax.bar(np.arange(len(y)) - bar_width/2, x1, bar_width, label='Group 1')
ax.bar(np.arange(len(y)) + bar_width/2, x2, bar_width, label='Group 2')

# Set title and labels
ax.set_title('Grouped Bar Chart Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_xticks(np.arange(len(y)))
ax.set_xticklabels(y)
ax.legend()

# Show the plot
plt.show()
```

## BAR CHART FOR CATEGORICAL VARIABLE (GENDER)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']}
df = pd.DataFrame(data)

# Count the frequency of each gender
gender_counts = df['Gender'].value_counts()

# Create a figure and axis
fig, ax = plt.subplots()

# Create a bar chart
ax.bar(gender_counts.index, gender_counts.values)

# Set title and labels
ax.set_title('Distribution of Genders')
ax.set_xlabel('Gender')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()
```

## HISTOGRAM FOR CONTINUOUS VARIABLE WITH MULTIPLE PEAKS(iq scores)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats

# Sample data
data = {'IQ Score': np.random.normal(100, 15, 100)}
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots()

# Create a histogram
ax.hist(df['IQ Score'], bins=20, edgecolor='black', alpha=0.5)

# Fit a normal distribution to the data
mean, std = stats.norm.fit(df['IQ Score'])
x = np.linspace(mean - 3*std, mean + 3*std, 100)
ax.plot(x, stats.norm.pdf(x, mean, std), 'r')

# Set title and labels
ax.set_title('Distribution of IQ Scores')
ax.set_xlabel('IQ Score')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()
```
